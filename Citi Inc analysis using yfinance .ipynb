# Citigroup Inc Cashflow

```python
# Here I am glabbing the net_income data from the dataframe and using it 
# to plot the bar.

net_income = go.Bar(
    x = ['2020', '2019', '2018'],
    y = cashflow.iloc[6],
    name = 'Net Income'
)
net_borrowing = go.Bar(
    x = ['2020', '2019', '2018'],
    y = cashflow.iloc[3],
    name = 'Net Borrowing'
)
capital_expediture = go.Bar(
    x = ['2020', '2019', '2018'],
    y = cashflow.iloc[16],
    name = 'Capital expediture'
)

# This part is where the plots are instanciated and set up for comparison 
# between the variables described above
data = [net_income, net_borrowing, capital_expediture]
layout = go.Layout(barmode='group')

fig = go.Figure(data=data, layout=layout)
iplot(fig, filename="Paul Plots")
```

# Citigroup Inc Balance Sheet

```python
balance_sheet = company.balance_sheet
balance_sheet
```

```python
# These are all the variable that I want to compare in the bar chart
total_assets = go.Bar(
    x = ['2020', '2019', '2018'],
    y = balance_sheet.iloc[6],
    name = 'Total Assets'
)
total_current_assets = go.Bar(
    x = ['2020', '2019', '2018'],
    y = balance_sheet.iloc[20],
    name = 'Total Current Assets'
)
total_current_liabilites = go.Bar(
    x = ['2020', '2019', '2018'],
    y = balance_sheet.iloc[15],
    name = 'Total Current Liabilites'
)
long_term_debt = go.Bar(
    x = ['2020', '2019', '2018'],
    y = balance_sheet.iloc[25],
    name = 'Long Term Debt'
)

inventory = go.Bar(
    x = ['2020', '2019', '2018'],
    y = balance_sheet.iloc[26],
    name = 'Inventory'
)
cash = go.Bar(
    x = ['2020', '2019', '2018'],
    y = balance_sheet.iloc[14],
    name = 'Cash'
)

# Creating the barchart
data = [total_assets, total_current_assets, total_current_liabilites, 
        long_term_debt, inventory, cash]
layout = go.Layout(barmode='group')

fig = go.Figure(data=data, layout=layout)
iplot(fig, filename="Paul Plots")
```
